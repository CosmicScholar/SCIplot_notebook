## 曲线图

```python
import matplotlib.pyplot as plt
import numpy as np
import sciplot 
import sciplot.figure as scifig
scifig.conf(fontname='Times New Roman',mathfont='cm')
style_plot=['paper']
fig=plt.figure(figsize=scifig.figsize('full',ratio='2:1'))
# 1. 生成信号
dt = np.pi / 100.
fs = 1. / dt
t = np.arange(0, 8, dt)
y = 10. * np.sin(2 * np.pi * 4 * t) + 5. * np.sin(2 * np.pi * 4.25 * t)
y = y + np.random.randn(*t.shape)
# 2. 绘制信号曲线
plt.plot(t, y)
# 3. 标注坐标轴
plt.xlabel('Time (s)')
plt.ylabel('Signal')
plt.tight_layout()
plt.savefig(figpath+'/chap1_lines.'+fig_fmt,dpi=fig_dpi)
plt.show()
```

## 蛛网图

```python
fig=plt.figure(figsize=scifig.figsize('full',ratio='2:1'))
plt.plot(t[1:30], y[1:30],'-o',ms=10,mfc='yellow',mec='red')
# 3. 标注坐标轴
plt.xlabel('Time (s)')
plt.ylabel('Signal')
plt.tight_layout()
plt.savefig(figpath+'/chap1_lines_SpiderWeb.'+fig_fmt,dpi=fig_dpi)
plt.show()
```

```python
fig=plt.figure(figsize=scifig.figsize('full',ratio='2:1'))
plt.plot(t, y,'-o',ms=10,mfc=(0,0,1,0.5),mew=0,markevery=10)
plt.plot(t, y,'.',mfc='orange')
# 3. 标注坐标轴
plt.xlabel('Time (s)')
plt.ylabel('Signal')
plt.tight_layout()
plt.savefig(figpath+'/chap1_lines_SpiderWeb_mkevery.'+fig_fmt,dpi=fig_dpi)
plt.show()
```

## 线型、线宽及颜色

```python
import matplotlib.pyplot as plt
import numpy as np
import sciplot 
import sciplot.figure as scifig
scifig.conf(fontname='Times New Roman',mathfont='cm')
style_plot=['paper']
fig=plt.figure(figsize=scifig.figsize('full',ratio='2:1'))
# 预设曲线颜色
lcs = ['black','red','green','blue']
# 预设线宽
lws=[1,2,3,4,5]
# 预设线型
ls=['-.','-',':','--']
# 生成不同相位的正弦信号
x = np.linspace(0, 2 * np.pi)
offsets = np.linspace(0, np.pi, 4, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
for i in range(0,len(ls)):
    plt.plot(x,yy[:,i],c=lcs[i],ls=ls[i],lw=lws[i],label=str('$\phi$ = %.2f $\pi$'% (offsets[i]/np.pi)))
plt.title('$y = sin(x + \phi)$')
# 显示图例
plt.legend()
plt.tight_layout()
plt.savefig(figpath+'/chap1_lines_style_width'+'.'+fig_fmt,dpi=fig_dpi)
plt.show()
```

## 自定义线型

```python
import matplotlib.pyplot as plt
import numpy as np
import sciplot 
import sciplot.figure as scifig
scifig.conf(fontname='Times New Roman',mathfont='cm')
style_plot=['paper']
fig=plt.figure(figsize=scifig.figsize('full',ratio='2:1'))
# 预设线宽
lws=[2,2,2,2,2,2,2]
# 预设线型
ls=[(0,(1,3)),(0,(1,5)),(0,(3,1)),(0,(3,3)),(0,(3,3,1,3)),(0,(3,3,1,3,1,3))]
# 生成不同相位的正弦信号
x = np.linspace(0, 2 * np.pi)
offsets = np.linspace(0, np.pi, 6, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
for i in range(0,len(ls)):
    plt.plot(x,yy[:,i],ls=ls[i],lw=lws[i],label=ls[i])
plt.title('$y = sin(x + \phi)$')
plt.ylim(-1.8,1.1)
# 显示图例
plt.legend(ncol=3,loc='lower center')
plt.tight_layout()
plt.savefig(figpath+'/chap1_lines_style_user'+'.'+fig_fmt,dpi=fig_dpi)
plt.show()
```
